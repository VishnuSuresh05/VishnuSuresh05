```python
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
total_sum=np.sum(arr)
print("total sum is ",total_sum)
column_sum=np.sum(arr,axis=0)
print(column_sum)

mean= np.mean(arr)
print("mean",mean)


max_value = np.max(arr,axis=1)
print("max value in ro",max_value)
min_value = np.min(arr)
print("min value in the array",min_value)
maxi_value= np.max(arr,axis=0)
print(maxi_value)

median = np.median(arr)
print(median)

median_col = np.median(arr,axis=1)
print(median_col)
```

```python
mat1 = np.random.randint(1,3,size=(3,3))
mat2= np.random.randint(1,5,size=(3,3))
print(mat1)
print(mat2)

matrix_multiplication = np.dot(mat1,mat2)
print("the result of matrix multiplication")
print(matrix_multiplication)

#calculate sum and mean

sum_matrix = np.sum(matrix_multiplication)
mean_matrix=np.mean(matrix_multiplication)

print("sum of the matrix is ",sum_matrix)
print("mean of the matrix is ",mean_matrix)

```
